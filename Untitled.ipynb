## Customer Segementation for better ads compaining

        This data set contains the following features:

        'User ID': unique identification for consumer

        'Age': cutomer age in years

        'Estimated Salary': Avg. Income of consumer

        'Gender': Whether consumer was male or female

        'Purchased': 0 or 1 indicated clicking on Ad

### Problem Defination:
    Find out the segment of customers who generally buys the products after clicking ad.

## Clustering

### 𝐇𝐢𝐞𝐫𝐚𝐫𝐜𝐡𝐢𝐜𝐚𝐥 𝐂𝐥𝐮𝐬𝐭𝐞𝐫𝐢𝐧𝐠

```python
hier_clust = linkage(df_std, method = 'ward')
```

```python
plt.figure(figsize = (12,9))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Observations')
plt.ylabel('Distance')
dendrogram(hier_clust,
           truncate_mode = 'level', 
           p = 5, 
           show_leaf_counts = False, 
           no_labels = True)
plt.show() 
```

Here we see that Data can be better divided into two segment

### Finding Best Number of Cluster with elbow method

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(df_std)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize = (10,8))
plt.plot(range(1, 11), wcss, marker = 'o', linestyle = '--')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('K-means Clustering')
plt.show()
```

### Kmeans Clustering

```python
kmeans = KMeans(n_clusters = 2, init = 'k-means++', random_state = 42)
kmeans.fit(df_std)
```

##  Segment Analysis

```python
df_segm_kmeans = df.copy()
df_segm_kmeans['Segment'] = kmeans.labels_
```

```python
df_segm_analysis = df_segm_kmeans.groupby(['Segment']).mean()
df_segm_analysis
```

```python
df_segm_kmeans.groupby(['Segment']).median()
```

* Both Segement has equal number of male and female
* Segment 0 customers do not purchase product at all
* Segement 0 customers are younger with low sallary
* Segement 1 97% customer buys the product
* Segement 1 customer are settled and with high sallary

```python
#Segment Renaming
df_segm_analysis.rename({0:'Non-Buyer',
                         1:'Buyer'})
```

```python
df_segm_kmeans['Segment Label']=df_segm_kmeans['Segment'].map({0:'Non-Buyer',
                         1:'Buyer'})
```

```python
# Segment Visualisation
plt.figure(figsize = (10, 8))
sns.scatterplot(df_segm_kmeans['Age'], df_segm_kmeans['EstimatedSalary'], hue = df_segm_kmeans['Segment Label'], 
                palette = ['r', 'g'])
plt.title('Segmentation K-means')
plt.show()
```
